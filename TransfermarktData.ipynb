```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

page = "https://www.transfermarkt.co.uk/1-bundesliga/torschuetzenliste/wettbewerb/L1/saison_id/2018"
pageTree = requests.get(page, headers=headers)
pageSoup = BeautifulSoup(pageTree.content, 'html.parser')
```

```python
Players = pageSoup.find_all("a", {"class": "spielprofil_tooltip"})

#Let's look at the first name in the Players list.
Players[0].text
```

```python
Goal = pageSoup.find_all("td", {"class": "zentriert hauptlink"})

Goal[0].text
```

```python
Age = pageSoup.find_all("td", {"class": ""})
```

```python
PlayersList = []
GoalList = []
AgeList = []

for i in range(0,25):
    PlayersList.append(Players[i+1].text)
    GoalList.append(Goal[i].text)
    AgeList.append(Age[i].text)
    
df = pd.DataFrame({"Players":PlayersList,"Goal":GoalList,"Age":AgeList})

df
```
